```python
whi = pd.read_csv('Happiness_Cantril_Ladder.csv')
```

```python
whi = whi.rename(columns={'World Happiness Report 2016 (Cantril Ladder (0=worst; 10=best))': 'WHIindex'})
whi.head()
```

```python
whi2016 = whi[whi['Year']==2016]
whi2015 = whi[whi['Year']==2015]
whi2014 = whi[whi['Year']==2014]
whi2013 = whi[whi['Year']==2013]
whi2012 = whi[whi['Year']==2012]
whi2011 = whi[whi['Year']==2011]
whi2010 = whi[whi['Year']==2010]
whi2009 = whi[whi['Year']==2009]
whi2008 = whi[whi['Year']==2008]
```

```python
worldaverageindex=whi
worldaverageindex.head()
```

```python
worldaverageindex = worldaverageindex.groupby(['Year'], as_index=False).mean()
```

```python
worldaverageindex
```

```python
worldaverageindex = worldaverageindex.drop([0,1,2])
```

```python
worldaverageindex = worldaverageindex.reset_index()
```

```python
worldaverageindex.drop(['index'], axis = 1, inplace = True)
```

```python
worldaverageindex
```

```python
years = worldaverageindex[['Year']]
index = worldaverageindex[['WHIindex']]
```

```python
years
```

```python
index
```

```python
plt.style.use('bmh')
plt.plot(years,(index/10)+0.3,'-o', c='navy', alpha=0.4)
plt.xlabel('Years')
plt.ylabel('World Happiness Index')
plt.title('Happiness Index Variation (2008-2016)')
plt.legend(['Average of Year'])
```
